```python
import model1_free_text_chunked_pipeline as m1fc
```

```python
train_df, test_df = m1fc.prepare_train_test_data(
    "free-text.csv",
    "synthetic_free_text.csv",
    chunk_size=15,
    test_size=0.2,
    random_state=42
)
```

```python
X_train, X_test, y_train, y_test = m1fc.preprocess_train_test(train_df, test_df)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
X_train_mi, X_test_mi, mi_features, mi_scores = m1fc.select_mutual_info(
    X_train, y_train, X_test, k=10
)
X_train_mi_scaled, X_test_mi_scaled = m1fc.scale_data(X_train_mi, X_test_mi)
mi_results = m1fc.evaluate_models(X_train_mi_scaled, X_test_mi_scaled, y_train, y_test)
```

```python
X_train_anova, X_test_anova, anova_features, anova_scores = m1fc.select_anova(
    X_train, y_train, X_test, k=10
)
X_train_anova_scaled, X_test_anova_scaled = m1fc.scale_data(X_train_anova, X_test_anova)
anova_results = m1fc.evaluate_models(X_train_anova_scaled, X_test_anova_scaled, y_train, y_test)
```

```python
X_train_corr, X_test_corr, corr_features, corr_scores = m1fc.select_correlation(
    X_train, y_train, X_test, k=10
)
X_train_corr_scaled, X_test_corr_scaled = m1fc.scale_data(X_train_corr, X_test_corr)
corr_results = m1fc.evaluate_models(X_train_corr_scaled, X_test_corr_scaled, y_train, y_test)
```

```python
X_train_var, X_test_var, var_features, var_scores = m1fc.select_variance(
    X_train, X_test, threshold=0.0, k=10
)
X_train_var_scaled, X_test_var_scaled = m1fc.scale_data(X_train_var, X_test_var)
var_results = m1fc.evaluate_models(X_train_var_scaled, X_test_var_scaled, y_train, y_test)
```

```python
X_train_md, X_test_md, md_features, md_scores = m1fc.select_mean_difference(
    X_train, y_train, X_test, k=10
)
X_train_md_scaled, X_test_md_scaled = m1fc.scale_data(X_train_md, X_test_md)
md_results = m1fc.evaluate_models(X_train_md_scaled, X_test_md_scaled, y_train, y_test)
```

```python
X_train_rfe_lr, X_test_rfe_lr, rfe_lr_features = m1fc.select_rfe_lr(
    X_train, y_train, X_test, k=10
)
X_train_rfe_lr_scaled, X_test_rfe_lr_scaled = m1fc.scale_data(X_train_rfe_lr, X_test_rfe_lr)
rfe_lr_results = m1fc.evaluate_models(X_train_rfe_lr_scaled, X_test_rfe_lr_scaled, y_train, y_test)
```

```python
X_train_rfe_svm, X_test_rfe_svm, rfe_svm_features = m1fc.select_rfe_svm(
    X_train, y_train, X_test, k=10
)
X_train_rfe_svm_scaled, X_test_rfe_svm_scaled = m1fc.scale_data(X_train_rfe_svm, X_test_rfe_svm)
rfe_svm_results = m1fc.evaluate_models(X_train_rfe_svm_scaled, X_test_rfe_svm_scaled, y_train, y_test)
```

```python
X_train_rfe_rf, X_test_rfe_rf, rfe_rf_features = m1fc.select_rfe_rf(
    X_train, y_train, X_test, k=10
)
X_train_rfe_rf_scaled, X_test_rfe_rf_scaled = m1fc.scale_data(X_train_rfe_rf, X_test_rfe_rf)
rfe_rf_results = m1fc.evaluate_models(X_train_rfe_rf_scaled, X_test_rfe_rf_scaled, y_train, y_test)
```

```python
X_train_sfm_lr, X_test_sfm_lr, sfm_lr_features, sfm_lr_scores = m1fc.select_sfm_lr(
    X_train, y_train, X_test, k=10
)
X_train_sfm_lr_scaled, X_test_sfm_lr_scaled = m1fc.scale_data(X_train_sfm_lr, X_test_sfm_lr)
sfm_lr_results = m1fc.evaluate_models(X_train_sfm_lr_scaled, X_test_sfm_lr_scaled, y_train, y_test)
```

```python
X_train_sfm_rf, X_test_sfm_rf, sfm_rf_features, sfm_rf_scores = m1fc.select_sfm_rf(
    X_train, y_train, X_test, k=10
)
X_train_sfm_rf_scaled, X_test_sfm_rf_scaled = m1fc.scale_data(X_train_sfm_rf, X_test_sfm_rf)
sfm_rf_results = m1fc.evaluate_models(X_train_sfm_rf_scaled, X_test_sfm_rf_scaled, y_train, y_test)
```
